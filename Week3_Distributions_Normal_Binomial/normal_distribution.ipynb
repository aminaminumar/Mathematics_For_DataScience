### Normal Distribution
- A normal distribution is a probability distribution that is suymmtric about the mean, showing that data near the mean are more frequent in occurence that data far from the mean.
### probability Distribution:
- A function thta shows the likelihood of different outcomes in an experiment

#### Lets import a data that is normally distributed

```python
#make necessary imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stat
```

```python
#load the data using pandas
data = pd.read_csv('normal_distribution_dataset.csv')
data.head()
```

```python
#plotting a histogram to check for normality
plt.figure(figsize = (10, 6))
sns.histplot(data['Height_cm'], kde=True)
plt.title('Histogram of Height')
plt.show()
```

```python
sns.histplot(data['Weight_kg'], kde=True, color='b')
plt.title('Weights of data')
plt.show()
```

```python
sns.histplot(data['Test_Score'], kde = True, color = 'r')
plt.title('TestScore')
plt.show()
```

```python
#plotting QQplot to check for noermality
stat.probplot(data['Height_cm'], dist="norm", plot=plt)
plt.title('QQ Plot for Height')
plt.show()
```
